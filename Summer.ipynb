# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('summer.csv')
data=[]
print("Summer Olympics is held so far in",len(df['City'].unique()),"cities")
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
medal=df[df['Medal']=='Gold']
medal=medal.groupby('Sport').count()['Medal'].sort_values(ascending=False).head()
print(medal)
medal.plot.bar(figsize =(10,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
sport=df.groupby('Sport').count()['Medal'].sort_values(ascending=False).head()
print(sport)
sport.plot.bar(figsize = (10, 5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
player = df.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head()
print(player)
player.plot.bar(figsize= (10,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
gold=df[df['Medal']=='Gold']
gold=gold.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head()
print(gold)
gold.plot.bar(figsize=(10,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
player=df[df['Medal']=='Gold']
year=player.sort_values(['Year'],ascending=True)
year=year[year['Country']=='IND'].iloc[0]
print(" India won first Gold Medal in",year['Year']," Summer Olympics")
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
popular=df.groupby('Event').count()['Athlete'].sort_values(ascending=False).head()
print(popular)
popular.plot.bar(figsize=(10,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
gold=df[df['Medal']=='Gold']
female=gold[df['Gender']=='Women']
female=female.groupby('Sport').count()['Medal'].sort_values(ascending=False).head()
print(female)
female.plot.bar(figsize=(10,5))
```
